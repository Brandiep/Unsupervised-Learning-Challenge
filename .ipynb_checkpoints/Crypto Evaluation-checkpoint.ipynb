```python
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
crypto_df = pd.read_csv(Path('crypto_data.csv'),index_col=0)
crypto_df.head()
```

### Clean the dataset

```python
#drop any coins that arent being traded
crypto_df = crypto_df[crypto_df['IsTrading'] != False]
crypto_df.head()
```

```python
#drop the 'is traded' column
crypto_df.drop('IsTrading', axis=1, inplace=True)
crypto_df.head()
```

```python
#drop rows with null values
crypto_df.isnull().sum()
```

```python
crypto_df = crypto_df.dropna()
crypto_df.head()
```

```python
#drop coins that have '0' mined
crypto_df = crypto_df[crypto_df['TotalCoinsMined'] != 0]
crypto_df.head()
```

```python
#delete coin name as it wont contribute to analysis
crypto_df.drop(['CoinName'], axis=1, inplace=True)
crypto_df.head()
```

```python
#convert all non numeric values to numbers
crypto_dummies = pd.get_dummies(crypto_df, columns=['Algorithm','ProofType'])
crypto_dummies.head()
```

- Using the get.dummies function takes the dataset from 4 columns to 100 columns

```python
#scale the dataset
scaler = StandardScaler().fit(crypto_dummies)
crypto_scaled=scaler.transform(crypto_dummies)
crypto_scaled
```

### PCA

```python
#dimentionally reduce the dataset using PCA
pca = PCA(n_components=0.9)
crypto_pca = pca.fit_transform(crypto_scaled)
```

```python
crypto_pca.shape
```

```python
# Fetch the explained variance
pca.explained_variance_ratio_.sum()
```

```python
# PCA Data loaded into dataframe
crypto_pca_df = pd.DataFrame(data=crypto_pca, index=crypto_df.index)
crypto_pca_df.head()
```

- using PCA we narrowed the dataset down from 100 columns to 75 columns

### t-SNE

```python
#reduce further with t-SNE
tsne = TSNE(learning_rate=35)
tsne_features = tsne.fit_transform(crypto_pca_df)
```

```python
tsne_features.shape
```

- t-SNE further reduces the dataset from 75 columns to only 2 columns

```python
# PCA Data loaded into dataframe
crypto_tsne_df = pd.DataFrame(data=tsne_features, index=crypto_df.index)
crypto_tsne_df.head()
```

```python
#Are there any dicernable clusters?
x = crypto_tsne_df[0]
y = crypto_tsne_df[1]

plt.scatter(x, y)
plt.show()
```

- based on the scatter plot, there seem to be a few notable clusters

### k-Means

```python
#create and elbow plot to determine best number of clusters

# Finding the best value for k
inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range of k values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(crypto_tsne_df)
    inertia.append(km.inertia_)

# Creating the Elbow Curve
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)

plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(list(range(11)))
plt.title('Elbow Curve')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
df_elbow
```

```python
# Initialize the k-means model
model = KMeans(n_clusters=4, random_state=0)

# Fit the model
model.fit(crypto_tsne_df)

# Predict clusters
predictions = model.predict(crypto_tsne_df)
```

```python
#visualize the clusters again
plt.scatter(x=tsne_features[:,0], y=tsne_features[:,1], c=model.labels_)
plt.xlabel('Principal component 1')
plt.ylabel('Principal component 2')
plt.title('Crypto clusters')
plt.colorbar()
plt.show()
```

## Analysis
- After dimentionally reducing the data set and using k-Means, we found that the coins can be clustered into 4 groups as shown by the elbow plot. To further showcase this, the data was also plotted and colorized based on the 4 model labels.
